### Gáfico de lineas

```python
x = [2, 3, 5, 6, 8]
y = [1, 5, 6, 8, 10]
plt.plot(x, y)
plt.grid()
```

```python
x = [2, 3, 5, 6, 8, 11]
y = [1, 5, 6, 8, 10, 3]

x2 = [1, 4, 7, 9, 10, 11]
y2 = [4, 9, 3, 12, 7, 5]

plt.plot(x, y, color='pink', linestyle='--', linewidth=5, label='Clase1')
plt.plot(x2, y2, color='orange', linestyle='-', linewidth=3, label='Clase2')
plt.legend(loc='upper left')
plt.grid()
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
```

```python
# Imprimir todas las semanas en un grafico de lineas en diferentes colores, con leyenda, titulos, etc 
productos =  ['camisa', 'pantalon', 'falda', 'bermudas']
semana1 = [150, 100, 200, 10]
semana2 = [80, 20, 50, 30]
semana3 = [160, 20, 120, 60]
semana4 = [10, 190, 30, 130]

plt.plot(productos, semana1, color='orange', linestyle='--', linewidth=3, label='semana1')
plt.plot(productos, semana2, color='red', linestyle='-', linewidth=3, label='semana2')
plt.plot(productos, semana3, color='pink', linestyle='--', linewidth=5, label='semana3')
plt.plot(productos, semana4, color='gold', linestyle='-', linewidth=5, label='semana4')
plt.legend(loc='upper left')
plt.grid()
plt.xlabel('PRODUCTOS')
plt.ylabel('Nº PRODUCTOS')
```

### Gráfico de barras

```python
# Diagramas de contorno
fig, ax = plt.subplots()
x = np.linspace(-3.0, 3.0, 100)
y = np.linspace(-3.0, 3.0, 100)
x, y = np.meshgrid(x, y)
z = np.sqrt(x**2 + 2*y**2)
ax.contourf(x, y, z)
plt.show()
```

```python
peliculas = {'spiderman': ['12-09-23', '23-06-24'], 'cars': [], 'enredados':['25-06-24', '30-09-23', '02-11-23']}
lista = {
}
import pandas as pd
def menu():
    print('-----------MENU--------------')
    print('1. Añadir película para ver')
    print('2. Añadir película ya vista')
    print('3. Añadir fecha a una película')
    print('4. Eleminar película')
    print('5. Mostrar listado de películas y sus fechas')
    print('6. Mostrar títulos de las películas')
    print('7. Mostrar peliculas sin ver')
    print('8. Mostrar todas las peliculas vistas dos veces')
    print('9. Salir y ver estadisticas')
```

```python
menu()
df=pd.DataFrame(lista)
fin = False
while fin == False:
    opcion = int(input('Escoge una opción del menú '))
    if opcion==1:
        peli = str(input('Nombre de la película para ver:')).lower()
        peliculas[peli]=[]
        print('\n:')
        del(df)
        df=pd.DataFrame(lista)
        df['Peliculas:']= peliculas.keys()
        df['Fechas:']= peliculas.values()
        print(df)
    elif opcion==2:
        peli = str(input('Nombre de la película ya vista:')).lower()
        fecha = str(input('Apunta en que fechas la viste:')).lower()
        peliculas[peli]=[fecha]
        del(df)
        df=pd.DataFrame(lista)
        print('\n')
        df['Peliculas:']= peliculas.keys()
        df['Fechas:']= peliculas.values()
        print(df)
    elif opcion==3:
        peli = str(input('Nombre de la película que quieres modificar:')).lower()
        fecha = str(input('Apunta las nuevas fechas:')).lower()
        peliculas[peli].append(fecha)
        del(df)
        df=pd.DataFrame(lista)
        print('\n')
        df['Peliculas:']= peliculas.keys()
        df['Fechas:']= peliculas.values()
        print(df)
    elif opcion==4:
        peli = str(input('Nombre de la película que quieres eliminar:')).lower()
        peliculas.pop(peli)
        del(df)
        df=pd.DataFrame(lista)
        print('\n')
        df['Peliculas:']= peliculas.keys()
        df['Fechas:']= peliculas.values()
        print(df)
    elif opcion==5:
        print('\n')
        df['Peliculas:']= peliculas.keys()
        df['Fechas:']= peliculas.values()
        print(df)
    elif opcion==6:
        print('\n')
       
        df['Peliculas:']= peliculas.keys()
        print(df['Peliculas:'])
    elif opcion==7:
        new_pelis = {}
        del(df)
        df=pd.DataFrame(lista)
        print(peliculas)
        for k, v in peliculas.items(): 
            if len(v) == 0: 
                new_pelis[k] = v
        df['Peliculas:']= new_pelis.keys()
        df['Fechas:']= new_pelis.values()   
        print(df)
    elif opcion==8:
        hola = {}
        del(df)
        df=pd.DataFrame(lista)
        for k, v in peliculas.items(): 
            if len(v) == 2: 
                hola[k] = v
        df['Peliculas:']= hola.keys()
        df['Fechas:']= hola.values()   
        print(df)
    
    elif (opcion == 9):
        print('Aquí están tus estadisticas')     
        fin = True
    

    else:
        print("Anda, segunda oportunidad, intenta otra vez a ver si nos enteramos.")
```
